```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2*np.pi,1000)
y=np.sin(x)
plt.plot(x,y)
plt.title('sine wave')
plt.xlabel('x')
plt.ylabel('sine(x)')
plt.grid(True)
plt.show()
```
